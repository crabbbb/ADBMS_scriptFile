```python
import pandas as pd

# set file path - 1000 record 
# csv_filePath = ["MOCK_DATA1.csv"
#                , "MOCK_DATA2.csv"
#                , "MOCK_DATA3.csv"
#                , "MOCK_DATA5.csv"]

# Filter MOCK_DATA4.csv - only 600 record ( 30 * 20 )
# Filter MOCK_DATA6.csv - only 280 record ( 14 * 20 )
csv_filePath = ["MOCK_DATA6.csv"]

# change variable 
numberOfRecordPickPerSection = 6

# ttlRecordExist = 1000
ttlRecordExist = 280 # per csv file 
splittingValue = 20 # every section have how many record 

# for calculate how many time need to loop per csv file 
value = ttlRecordExist // splittingValue

# for reading input.csv use 
startRow = 0 # always start at 0 ( dont change )
endRow = splittingValue - 1

# for generate new dataframe ( use for output.csv )
nStartRow = 0 # always start at 0 ( dont change )
nEndRow = numberOfRecordPickPerSection - 1
```

```python
# function declare for pick randam
def pickRandom(startRow, endRow):
    return df.loc[startRow:endRow].sample(n=numberOfRecordPickPerSection)
```

```python
# first = True # have header and overwirte file 
first = False

for filePath in csv_filePath:
    df = pd.read_csv(filePath)

    # create new dataframe for holding data and after finish select will be used for generate output.csv file 
    newDf = pd.DataFrame(data=None, columns=df.columns)

    print("\n")
    print(filePath)
    
    # for reading input.csv use 
    startRow = 0 # always start at 0 ( dont change )
    endRow = splittingValue - 1
    
    # for generate new dataframe ( use for output.csv )
    nStartRow = 0 # always start at 0 ( dont change )
    nEndRow = numberOfRecordPickPerSection - 1
    
    # looping for random selecting data 
    for x in range(value):
        
        # use pickRandom() to randomly select and insert to new dataframe
        newDf = newDf._append(pickRandom(startRow=startRow, endRow=endRow))
        
        # print(pickRandom(startRow=startRow, endRow=endRow))
        
        # update value 
        startRow = startRow + 20
        endRow = endRow + 20 
    
        nStartRow = nStartRow + 6
        nEndRow = nEndRow + 6

    print(newDf.count())
    
    # check first time 
    if (first):
        # export dataframe to csv file 
        newDf.to_csv('output1.csv', mode='w', index=False) # mode w = overwrite the file 
        first = False
    else:
        # export dataframe to csv file 
        newDf.to_csv('output1.csv', mode='a', index=False, header=False) # mode a = append the data to existing file 

```
